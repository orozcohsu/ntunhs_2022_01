```python
#create table schema
#Import the package pymysql for making a connection to the database
import pymysql

connection=pymysql.connect(host='172.105.215.62',
                         user='db',
                         password='20220306!',
                         db='test123')
```

```python
connection
```

```python
# create cursor
cursor=connection.cursor()
```

```python
# create the Customer Table
cursor.execute("""CREATE TABLE afterETL
    (
     ID INT NOT NULL,
     CustomerID varchar(5) NOT NULL,
     EmployeeID INT NOT NULL,
     ShipName varchar(40) NOT NULL,
     ShipCountry varchar(15) NOT NULL,
     OrderID INT NOT NULL,
     Freight FLOAT NOT NULL,
     Qty FLOAT NOT NULL,
     PRIMARY KEY (ID)
     )
     """)
connection.commit()
```

```python
#rename column name
df5=df5.rename({'index': 'ID'}, axis=1)
```

```python
# column name list
cols="`,`".join([str(i) for i in df5.columns.tolist()])

# Insert DataFrame recrds one by one.
for i,row in df5.iterrows():
    sql = "INSERT INTO `afterETL` (`" +cols + "`) VALUES (" + "%s,"*(len(row)-1) + "%s)"
    cursor.execute(sql, tuple(row))

    # the connection is not autocommitted by default, so we must commit to save our changes
    connection.commit()
```
