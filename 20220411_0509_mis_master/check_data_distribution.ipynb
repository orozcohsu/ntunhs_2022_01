```python
from scipy import stats
import numpy as np
import seaborn as sns
```

```python
a=np.random.gamma(6,size=200)
```

```python
#sns.distplot(a,fit=stats.gamma)
sns.distplot(a,fit=stats.gamma,kde=False)
```

```python
b=np.random.beta(7,7,size=200)
```

```python
sns.distplot(b,fit=stats.beta)
```

```python
c=np.random.normal(0,1,size=20000)
```

```python
sns.distplot(c,fit=stats.norm)
```
