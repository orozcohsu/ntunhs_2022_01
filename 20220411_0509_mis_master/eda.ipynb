```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import missingno
```

```python
#將 train.csv.zip 解壓縮為 train.csv

df = pd.read_csv('train.csv',low_memory=False)
df.info()
df.head()
```

```python
#1017209筆資料，9個欄位
len(df)
```

```python
#修改欄位型態
df['Store'] = df['Store'].astype('str')

df['DayOfWeek'] = df['DayOfWeek'].astype('category')
df['Open'] = df['Open'].astype('category')
df['Promo'] = df['Promo'].astype('category')
df['StateHoliday'] = df['StateHoliday'].astype(str).str.strip().astype('category')
df['SchoolHoliday'] = df['SchoolHoliday'].astype('category')

df['Date'] = pd.to_datetime(df['Date'])
```

```python
#修改欄位型態後，有節省記憶體使用空間
df.info()
```

```python
#檢查遺缺值
#刻意刪除一些資料，產生空欄位值 (Sales,Customers)

if df.isnull().any(axis=None):
    print("\nPreview of data with null values:\n")
    print(df[df.isnull().any(axis=1)].head(100))
    missingno.matrix(df)
    plt.show()
```

```python
#檢查欄位是否有空值
df.isnull().any()
```

```python
#檢查重複資料
#刻意做一些重複的資料(row)

if len(df[df.duplicated()]) > 0:
    print("No. of duplicated entries: ", len(df[df.duplicated()]))
    #列印出所有重複的資料
    print(df[df.duplicated(keep=False)].sort_values(by=list(df.columns)).head())
else:
    print("No duplicated entries found")
```

```python
#刪除重複資料

df.drop_duplicates(inplace=True)
```

```python
#再次檢查還有沒有重複的資料

if len(df[df.duplicated()]) > 0:
    print("No. of duplicated entries: ", len(df[df.duplicated()]))
    print(df[df.duplicated(keep=False)].sort_values(by=list(df.columns)).head())
else:
    print("No duplicated entries found")
```

```python
#進行category資料類型EDA
#nunique函數可以找出欄位的不重複值數量

def categorical_eda(df):
    print("檢視類別型欄位")
    print(df.select_dtypes(include=['category']).nunique())

    #進行繪圖(catplot)
    for col in df.select_dtypes(include='category').columns:
        #catplot圖表專門繪製categorical data
        fig = sns.catplot(x=col, kind="count", data=df)
        fig.set_xticklabels(rotation=90)
        plt.show()
```

```python
categorical_eda(df)
```

```python
#進行數值類型EDA

def numeric_eda(df, hue=None):
    print("檢視數值型欄位")
    display(df.describe().T)
    columns = df.select_dtypes(include=np.number).columns
    
    #參考講義add_subplot說明
    figure = plt.figure(figsize=(20, 10))
    #給第一個欄位繪圖
    figure.add_subplot(1, len(columns), 1)
    
    for index, col in enumerate(columns):
        if index > 0:
            #隨著欄位增加index順序持續加1
            figure.add_subplot(1, len(columns), index + 1)
        #boxplot為盒鬚圖
        sns.boxplot(y=col, data=df, boxprops={'facecolor': 'None'})
    #繪圖自動校正對齊
    figure.tight_layout()
    plt.show()
    
    #確定有類別型欄位
    if len(df.select_dtypes(include='category').columns) > 0:
        #選擇數值型欄位，目的是找出y軸的欄位
        for col_num in df.select_dtypes(include=np.number).columns:
            #選擇類別型欄位，目的是找出x軸欄位
            for col in df.select_dtypes(include='category').columns:
                #種類為小提琴圖
                #強化資料分佈與機率密度
                #aspect:圖片顯示的擁擠程度
                fig = sns.catplot(x=col, y=col_num, kind='violin', data=df, height=5, aspect=4)
                fig.set_xticklabels(rotation=90)
                plt.show()
    
    print("\n 進行 pairwise joint distribution 繪圖")
    #hue:類別型資料種類(通常用來圖表比對分析)
    #pairplot僅針對數值欄位進行
    if hue==None:
        sns.pairplot(df.select_dtypes(include=np.number))
    else:
        sns.pairplot(df.select_dtypes(include=np.number).join(df[[hue]]), hue=hue)
    plt.show()
```

```python
numeric_eda(df)
```

```python
#進行數值類型EDA(時間序列)

def time_series_plot(df):
    print("針對數值型欄位by daily, monthly and yearly頻次繪圖")
    #選擇日期時間欄位
    if len(df.select_dtypes(include='datetime64').columns)>0:
        for col in df.select_dtypes(include='datetime64').columns:
            #定義有三種繪圖類型
            for p in ['D', 'M', 'Y']:
                if p=='D':
                    print("Plotting daily data")
                elif p=='M':
                    print("Plotting monthly data")
                else:
                    print("Plotting yearly data")
                for col_num in df.select_dtypes(include=np.number).columns:
                    #將資料集複製
                    __ = df.copy()
                    #將col欄位設定為資料集的index欄位，col為含有日期型態的資料
                    __ = __.set_index(col)
                    #以D,M,Y等日期計算方式，將剩餘的數值型欄位進行加總
                    __T = __.resample(p).sum()
                    #資料集直接畫圖，方法為plot
                    ax = __T[[col_num]].plot()
                    #設定畫圖的y軸起始點為0開始
                    ax.set_ylim(bottom=0)
                    #讓y軸的數字不要用科學符號呈現
                    ax.get_yaxis().set_major_formatter(
                        matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
                    plt.show()
```

```python
time_series_plot(df)
```
